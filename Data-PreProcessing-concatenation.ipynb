```python
# if many columns have null values we use imputer
def quanqual(dataset):
    qual=[]
    quan=[]
    for columnname in dataset.columns:
        print(columnname)
        if(dataset[columnname].dtype=='O'):
            print('qual')
            qual.append(columnname)
        else:
            print('quan') 
            quan.append(columnname)
    return qual,quan
```

```python
quanqual(dataset)
```

```python
qual,quan=quanqual(dataset)
```

```python
quan
```

```python
qual
```

```python
# using imputer for dealing with many columns with null values
import numpy as np
from sklearn.impute import SimpleImputer
imp=SimpleImputer(missing_values=np.nan,strategy='mean')
imp.fit(dataset[quan])
dataset_quan=imp.transform(dataset[quan])
```

```python
dataset_quan=pd.DataFrame(dataset,columns=quan)
```
